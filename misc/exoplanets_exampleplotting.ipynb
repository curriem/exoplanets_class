```julia
using Pkg
Pkg.add("PyPlot")
using PyPlot
```

```julia
x = range(0,stop=2*pi*4,length=1000);
ecc = 0.999;
y = x - ecc * sin.(x) ;
plot(y, x, color="red", linewidth=10.0);
plot(y,y);
ylabel("E"); 
xlabel("M"); 
title("Kepler's equation for e=0.999");
annotate(L"M = E - e \sin{E}",xy=[1;25],fontsize=20.0);
```
