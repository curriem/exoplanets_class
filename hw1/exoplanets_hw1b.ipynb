# Periodogram

```julia
function get_phase(time, P)
    phase = (time ./ P) .% 1
    return phase
end;
```

Define the "scatter" between RV measurements **sorted by phase** at period P as
$$
S_P = \sum_{i = 1}^{i = N - 1} (v_{i} - v_{i + 1})^2
$$

```julia
function compute_scatter(rv_by_phase)
    N = size(rv_by_phase)[1]
    scatter = 0
    for i in range(1, stop=N - 1)
        scatter = scatter + (rv_by_phase[i] - rv_by_phase[i + 1])^2
    end
    return scatter
end;
```

```julia
function periodogram(rv_data, times_data, P_vals)
    scatters_list = []
    for period in P_vals
        phase = get_phase(times_data, period)
        phase_sorted_inds = sortperm(phase)
        rv_by_phase = rv_data[phase_sorted_inds]
        rv_scatter = compute_scatter(rv_by_phase)
        append!(scatters_list, rv_scatter)
    end
    return scatters_list
end;
```

```julia
function plot_folded_data(period)
    figure();
    phase = get_phase(time_data, period)
    errorbar(phase, rv_data, yerr=rv_err, fmt="o", markersize=1, ecolor="k", elinewidth=1, capsize=2, barsabove=true)#, markersize=0, ecolor="k")
    xlabel("Phase")
    ylabel("Radial Velocity (km/s)")
end;

function plot_model(period, e, t_p)
    phase = get_phase(time_data, period)
    
    M_vals = M(period, time_data, t_p)
    E_vals = kepler_solver(M_vals, e)
    f_vals = f(e, E_vals)
    
    h, c, v0, _ = optim_h_c_v0(period, e, t_p, rv_data, time_data, rv_err)
    
    idxs = sortperm(phase)
    plot(phase[idxs], model_rv(h, c, v0, f_vals)[idxs])
end;

function plot_residuals(period, e, t_p)
    phase = get_phase(time_data, period)
    
    M_vals = M(period, time_data, t_p)
    E_vals = kepler_solver(M_vals, e)
    f_vals = f(e, E_vals)
    
    h, c, v0, _ = optim_h_c_v0(period, e, t_p, rv_data, time_data, rv_err)
    
    idxs = sortperm(phase)
    figure()
    errorbar(phase[idxs], rv_data[idxs] .- model_rv(h, c, v0, f_vals)[idxs], yerr=rv_err, fmt="o", markersize=1, ecolor="k", elinewidth=1, capsize=2, barsabove=true)
    xlabel("Phase")
    ylabel("Model Residual (km/s)")
end;
```

```julia
P_grid = range(1, stop=200, length=10000)
scatters = periodogram(rv_data, time_data, P_grid)

figure()
plot(P_grid, scatters)
xlabel("Period (JD)")
ylabel("Scatter")
show();

best_P = P_grid[sortperm(scatters)[1]]

plot_folded_data(best_P)

print("Minimum scatter at P = ")
print(best_P)
```

# Optimize over orbital parameters

$ M = \frac{2\pi}{P}(t - t_p) $

`E = newton_method(M, e)`

$ f = 2 \mathrm{arctan}[(\frac{1+e}{1-e})^{1/2} \mathrm{tan}(\frac{E}{2})] $

```julia
function model_rv(h, c, v0, f)
   return h .* cos.(f) .+ c .* sin.(f) .+ v0 
end;
```

```julia
function chi_sq(rv_data, rv_err, model_rvs)
    return sum((rv_data .- model_rvs).^2 ./ rv_err.^2)
end;
```

```julia
function dgdE(E, e)
    return 1 - e*cos(E)
end;
```

```julia
function g(E, e, M)
    return E - e*sin(E) - M
end;
```

```julia
# Check the convergence criterion of the kepler solver
function check_convergence(conv_arr)
    for a in conv_arr
        if a > 1E-9
           return false 
        end
    end
    return true
end

# Compute E for given M and e
function kepler_solver(M, e)
    E0 = M + 0.85 * e * sign.(sin.(M))
    g0 = g.(E0, e, M)
    dgdE0 = dgdE.(E0, e)
    converged = false
    i = 0
    max_iter = 100
    while !converged
        E = E0 - g.(E0, e, M) ./ dgdE.(E0, e)
        E0 = E
        converged = check_convergence(abs.(-g.(E0, e, M) ./ dgdE.(E0, e)))
        i = i + 1
        if i > max_iter
            return E0
        end
    end
    return E0
end;
```

```julia
# Compute f from e and E
function f(e, E)
    return 2*atan.( ((1 + e) / (1 - e))^2 * tan.(E/2))
end;
```

```julia
# Compute M from P, t, t_p, where t are the times associated with RV measurements
function M(P, t, t_p)
    return 2 * pi / P * (t .- t_p)
end;
```

# steps:

1) given: P, e, t_p: optimize for these vals

2) solve for M over all measurements:

$ M = \frac{2\pi}{P} ( t - t_p ) $

3) Solve for E over all measurements

$ M = E - e \sin(E) $

`E = newton(M, e)`

4) Solve for f over all measurements

$ f = 2 \tan^{-1} ( ( \frac{1 + e}{1 - e} )^{1/2} \tan(E / 2) ) $

5) Solve analytically for $h$, $c$, $v_0$ that minimize $\chi^2$

The following code implements the procedure outlined by [Wright & Howard (2009)](https://iopscience.iop.org/article/10.1088/0067-0049/182/1/205/pdf).

```julia
# Compute W given the uncertainties on the RV measurements
# W = \delta_ij sigma_i^2
# Where sigma_i is the RV uncertainty
function init_W(rv_errs)
    return Diagonal(1 ./ rv_errs.^2)
end;
```

```julia
# Compute F, the matrix containing coefficents of the linear parameters
function init_F(f)
    N = size(f)[1]
    cos_f = cos.(f)
    sin_f = sin.(f)
    ones_f = ones(N)
    
    F = transpose(hcat(cos_f, sin_f, ones_f))
    return F
end;
```

```julia
# Compute epsilon, the variance-covariance matrix given F and W
function compute_eps(F, W)
    N = size(F)[1]
    little_bit_of_identity = 1E-9 .* Diagonal(ones(N))
    eps = inv(F * W * transpose(F) + little_bit_of_identity)
    return eps
end;
```

```julia
# Compute the values of the linear paramters given F, W, epsilon, and the RV measurements
function compute_beta(F, W, eps, rv_data)
    beta = reshape(rv_data, (1, size(rv_data)[1])) * W * transpose(F) * eps
    return beta
end;
```

```julia
# For given P, e, t_p and RV measurements, associated times and errors, compute the values of
# h, c, and v_0 which minimize the chi-squared between Keplerian model and measured RVs
function optim_h_c_v0(P, e, t_p, rv_data, time_data, rv_err)
    M_vals = M(P, time_data, t_p)
    E_vals = kepler_solver(M_vals, e)
    f_vals = f(e, E_vals)
    
    W = init_W(rv_err)
    F = init_F(f_vals)
    eps = compute_eps(F, W)
    beta = compute_beta(F, W, eps, rv_data)
    
    h, c, v0 = beta[1], beta[2], beta[3]
    
    chi_squared = chi_sq(rv_data, rv_err, model_rv(h, c, v0, f_vals))
        
    return h, c, v0, chi_squared
end;
```

```julia
# Function to use with the Optim package to optimize over the non-linear parameters
# P, e, and t_p
function optim_P_e_t_p(x)
    P, e, t_p = x[1], x[2], x[3]
    # Perform bounds checking
    if P < 0 || t_p < 0 || e < 0 || e > 1
        return 1E14
    end
    h, c, v0, chi_squared = optim_h_c_v0(P, e, t_p, rv_data, time_data, rv_err)
    return chi_squared
end;
```

Non-linear optimization over $P$, $e$, and $t_p$

```julia
function compute_gamma(h, v0, e)
   return v0 - e * h 
end;

function compute_K(h, c)
   return sqrt(h^2 + c^2) 
end;

function compute_omega(h, K)
    return acos(h / K)
end;
```

We perform an optimization over $P$, $e$, and $t_p$ using the gradient free Nelder-Mead algorithm with *particularly* good initial guesses for $P$, $e$, and $t_p$.

```julia
res = optimize(optim_P_e_t_p, [best_P, 0.9, 2454424.857])
Optim.summary(res)

min_res = Optim.minimizer(res)
P, e, t_p = min_res[1], min_res[2], min_res[3]
h, c, v0, chi_squared = optim_h_c_v0(P, e, t_p, rv_data, time_data, rv_err)

plot_folded_data(P)
plot_model(P, e, t_p)
plot_residuals(P, e, t_p)
show();

gam = compute_gamma(h, v0, e)
kam = compute_K(h, c)
wam = compute_omega(h, kam)

print("P = ", P, "\n")
print("e = ", e, "\n")
print("t_p = ", t_p, "\n")
print("gamma = ", gam, "\n")
print("K = ", kam, "\n")
print("omega = ", wam, "\n")
print("Chi-Squared = ", chi_squared, "\n")
```

The above method uses the gradient free Nelder-Mead algorithm through Julia's Optim package. Given specific initial values close to the known values, we recover the correct HD 80606b $P$, $e$, and $t_p$ parameters. Below, we perform an alternate method, optimizing over a grid of $e$ and $t_p$ values. Here, we find a much different $e$ and we find $t_p$ to be off by $\sim$one period. Despite the method above giving a better value for $e$, the method below gives a more realistic value for both $\gamma$ and $K$. Searching the grid is more similar to the approach we would take if we had not known the true values for this planet. 

```julia
min_e, min_t_p, min_chi_squared = 1E10, 1E10, 1E10
for e in range(0, stop=1, length=100)
    # The time of periastron passage must be between any time and that time plus the period
    # Using the last day observed + the period gives a little more wiggle room
    for t_p in range(minimum(time_data), stop=maximum(time_data) + best_P, length=100)
        P = best_P
        chi_squared = try
        h, c, v0, chi_squared = optim_h_c_v0(P, e, t_p, rv_data, time_data, rv_err)
            chi_squared
        catch
            1E10
        end
        if chi_squared < min_chi_squared
            min_chi_squared = chi_squared
            min_e = e
            min_t_p = t_p
        end
    end
end

h, c, v0, chi_squared = optim_h_c_v0(best_P, min_e, min_t_p, rv_data, time_data, rv_err)

gam = compute_gamma(h, v0, min_e)
kam = compute_K(h, c)
wam = compute_omega(h, kam)

print("P = ", best_P, "\n")
print("e = ", min_e, "\n")
print("t_p = ", min_t_p, "\n")
print("gamma = ", gam, "\n")
print("K = ", kam, "\n")
print("omega = ", wam, "\n")
print("Chi-Squared = ", chi_squared, "\n")

plot_folded_data(best_P)
plot_model(best_P, min_e, min_t_p)
show();
plot_residuals(best_P, min_e, min_t_p)
show();
```
